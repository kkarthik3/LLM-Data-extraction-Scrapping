```python
import pandas as pd
```

```python
df = pd.read_excel("scraped_2511.xlsx")
```

```python
df = df.dropna(axis=0)
```

```python
df.shape
```

```python
df.head()
```

```python
import tiktoken

def tokenier_couont(text):
    encoding = tiktoken.get_encoding("gpt2")
    tokens = encoding.encode(text)
    token_length = len(tokens)
    return token_length
```

```python
df["Token_count"] = df["All_content"].apply(tokenier_couont)
```

```python
df["Token_count"].sum()
```

```python
filtered_df = df[~df["All_content"].str.contains("Technical issue", case=False, na=False)]
filtered_df
```
